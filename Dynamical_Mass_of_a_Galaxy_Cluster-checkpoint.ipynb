```python
# Step 1: Import Required Libraries and Constants
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy.constants import G, c
from astropy.cosmology import Planck18 as cosmo

# Constants
c_val = c.value                 # Speed of light in m/s
G_val = G.value                 # Gravitational constant in m^3 kg^-1 s^-2
```

```python
# Step 2: Load Cleaned Galaxy Cluster CSV
df = pd.read_csv(r"C:\Users\suman\Downloads\cluster_members.csv")
```

```python
# Step 3: Remove duplicates using objid and average data
averaged_df = df.groupby('objid').agg({
    'specz': 'mean',
    'ra': 'first',
    'dec': 'first',
    'proj_sep': 'first'
}).reset_index()
```

```python
# Step 4: Redshift Filtering (3-sigma clipping)
z_mean = averaged_df['specz'].mean()
z_std = averaged_df['specz'].std()
filtered_df = averaged_df[(averaged_df['specz'] >= z_mean - 3*z_std) & (averaged_df['specz'] <= z_mean + 3*z_std)]
```

```python
# Step 5: Compute Cluster Redshift and Velocity Dispersion
z_cluster = filtered_df['specz'].mean()
velocities = c_val * (filtered_df['specz'] - z_cluster) / (1 + z_cluster)
dispersion = np.std(velocities)
disp_kms = dispersion / 1000

print(f"üìå Cluster redshift (z): {z_cluster:.5f}")
print(f"üìå Velocity dispersion: {disp_kms:.2f} km/s")

if disp_kms > 2000:
    print("‚ö†Ô∏è Unusually high velocity dispersion detected.")
```

```python
# Step 6: Estimate Angular Radius and Physical Radius (correct way)
theta_arcmin = filtered_df['proj_sep'].quantile(0.9)   # 90th percentile angular radius
theta_rad = np.deg2rad(theta_arcmin / 60)              # convert arcmin to radians

# ‚úÖ Use correct cosmological angular diameter distance from Astropy
d_ang_mpc = cosmo.angular_diameter_distance(z_cluster).value  # in Mpc

# Physical radius in Mpc
radius_mpc = d_ang_mpc * theta_rad
diameter_mpc = radius_mpc * 2
print(f"üìè Cluster physical diameter: {diameter_mpc:.3f} Mpc")

if diameter_mpc > 5:
    print("‚ö†Ô∏è Cluster diameter unusually large. Typical clusters are 1‚Äì3 Mpc.")
```

```python
# Step 7: Calculate Dynamical Mass
radius_m = radius_mpc * 3.086e22   # Mpc to meters
disp_mps = dispersion              # already in m/s

# Virial theorem: M = 3 * sigma^2 * R / G
M_dyn_kg = 3 * (disp_mps ** 2) * radius_m / G_val
M_dyn_solar = M_dyn_kg / 1.989e30

print(f"üåå Estimated Dynamical Mass = {M_dyn_solar:.2e} solar masses")
```

```python
# Step 8: Additional Visualizations for Analysis

import seaborn as sns
import matplotlib.ticker as ticker

# --- Plot 1: Histogram of Redshifts (after filtering)
plt.figure(figsize=(8, 4))
plt.hist(filtered_df['specz'], bins=25, color='skyblue', edgecolor='black')
plt.axvline(z_cluster, color='red', linestyle='--', label=f'Mean z = {z_cluster:.5f}')
plt.title("Histogram of Galaxy Redshifts")
plt.xlabel("Spectroscopic Redshift (z)")
plt.ylabel("Galaxy Count")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Plot 2: Velocity Distribution Histogram
velocities = c_val * (filtered_df['specz'] - z_cluster) / (1 + z_cluster)
dispersion = np.std(velocities)
velocities_kms = velocities / 1000

plt.figure(figsize=(8, 4))
plt.hist(velocities_kms, bins=25, color='lightcoral', edgecolor='black')
plt.axvline(0, color='black', linestyle='--')
plt.axvline(disp_kms, color='blue', linestyle='--', label=f'+œÉ = {disp_kms:.1f} km/s')
plt.axvline(-disp_kms, color='blue', linestyle='--', label=f'-œÉ = {-disp_kms:.1f} km/s')
plt.title("Galaxy Velocity Distribution (Relative to Cluster z)")
plt.xlabel("Peculiar Velocity (km/s)")
plt.ylabel("Galaxy Count")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Plot 3: Angular Separation Histogram
plt.figure(figsize=(8, 4))
plt.hist(filtered_df['proj_sep'], bins=20, color='lightgreen', edgecolor='black')
plt.axvline(theta_arcmin, color='purple', linestyle='--', label=f'90th %ile = {theta_arcmin:.2f} arcmin')
plt.title("Projected Angular Separation from Cluster Center")
plt.xlabel("Angular Separation (arcmin)")
plt.ylabel("Number of Galaxies")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Plot 4: Spatial Distribution RA vs DEC
plt.figure(figsize=(6, 6))
plt.scatter(filtered_df['ra'], filtered_df['dec'], c=filtered_df['specz'], cmap='viridis', s=20, edgecolor='black')
plt.colorbar(label='Redshift')
plt.title("Spatial Distribution of Cluster Galaxies")
plt.xlabel("Right Ascension (RA)")
plt.ylabel("Declination (Dec)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Plot 5: Velocity vs Projected Separation
plt.figure(figsize=(8, 5))
plt.scatter(filtered_df['proj_sep'], velocities_kms, alpha=0.7, color='teal', edgecolor='k')
plt.axhline(disp_kms, color='red', linestyle='--', label='+œÉ')
plt.axhline(-disp_kms, color='red', linestyle='--', label='-œÉ')
plt.axhline(0, color='black', linestyle='--')
plt.title("Galaxy Peculiar Velocity vs Projected Separation")
plt.xlabel("Projected Separation (arcmin)")
plt.ylabel("Peculiar Velocity (km/s)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
